### Create base schedule

```python
base_schedule_data = []

for emp_id in employee_df['employee_id']:
    for i, date in enumerate(schedule_dates):
        # Determine working status based on 2x2 pattern
        is_working = (i % 4) < 2
        status = 'working' if is_working else 'vacation'
        base_schedule_data.append({'date': date, 'employee_id': emp_id, 'base_status': status})

base_schedule_df = pd.DataFrame(base_schedule_data)
display(base_schedule_df.head())
display(base_schedule_df.tail())
```

### Add extra vacation days

#### Subtask:
Distribute the 24 additional vacation days per employee, ensuring they are added to the base vacation days.


**Reasoning**:
I need to add 24 additional vacation days to the base schedule for each employee. I will do this by randomly selecting 24 'working' days for each employee and changing their status to 'vacation' in a new column to preserve the original base schedule.



```python
employee_schedule_df = base_schedule_df.copy()
employee_schedule_df['final_status'] = employee_schedule_df['base_status']

additional_vacation_days_per_employee = 24

for emp_id in employee_schedule_df['employee_id'].unique():
    # Get the indices of the working days for the current employee
    working_days_indices = employee_schedule_df[(employee_schedule_df['employee_id'] == emp_id) & (employee_schedule_df['base_status'] == 'working')].index

    # Randomly select 24 indices from the working days
    if len(working_days_indices) >= additional_vacation_days_per_employee:
        additional_vacation_indices = np.random.choice(working_days_indices, additional_vacation_days_per_employee, replace=False)
        # Update the final_status for the selected indices to 'vacation'
        employee_schedule_df.loc[additional_vacation_indices, 'final_status'] = 'vacation'
    else:
        # If there are fewer than 24 working days, convert all of them to vacation
        employee_schedule_df.loc[working_days_indices, 'final_status'] = 'vacation'


display(employee_schedule_df.head())
display(employee_schedule_df.tail())
```

### Ensure minimum coverage

#### Subtask:
For each day, check if the minimum number of employees are scheduled for each warehouse (3 for small, 4 for medium, 7 for large). Adjust the schedule if necessary to meet these minimums, prioritizing working days over extra vacation days if needed.


**Reasoning**:
Merge the employee schedule with employee data to get the assigned warehouse and then iterate through each date and warehouse to check and adjust the schedule to meet minimum employee coverage.



```python
employee_schedule_df = pd.merge(employee_schedule_df, employee_df[['employee_id', 'assigned_warehouse']], on='employee_id', how='left')

# Define minimum coverage
min_coverage = {1: 3, 2: 4, 3: 7}

# Iterate through each date and warehouse
for date in employee_schedule_df['date'].unique():
    for warehouse_id in employee_schedule_df['assigned_warehouse'].unique():
        # Filter schedule for current date and warehouse
        daily_warehouse_schedule = employee_schedule_df[(employee_schedule_df['date'] == date) & (employee_schedule_df['assigned_warehouse'] == warehouse_id)].copy()

        # Count working employees
        working_employees_count = daily_warehouse_schedule[daily_warehouse_schedule['final_status'] == 'working'].shape[0]

        # Get required minimum coverage
        required_coverage = min_coverage.get(warehouse_id, 0) # Default to 0 if warehouse_id not in dict

        # Adjust schedule if below minimum
        if working_employees_count < required_coverage:
            employees_on_vacation = daily_warehouse_schedule[daily_warehouse_schedule['final_status'] == 'vacation']

            # Prioritize converting extra vacation days back to working
            extra_vacation_candidates = employees_on_vacation[employees_on_vacation['base_status'] == 'working']

            num_to_convert = required_coverage - working_employees_count

            # Convert extra vacation days first
            if extra_vacation_candidates.shape[0] >= num_to_convert:
                indices_to_convert = np.random.choice(extra_vacation_candidates.index, num_to_convert, replace=False)
                employee_schedule_df.loc[indices_to_convert, 'final_status'] = 'working'
            else:
                # Convert all extra vacation days and then some base vacation days if needed
                indices_to_convert_extra = extra_vacation_candidates.index
                employee_schedule_df.loc[indices_to_convert_extra, 'final_status'] = 'working'
                converted_count = indices_to_convert_extra.shape[0]
                remaining_to_convert = num_to_convert - converted_count

                if remaining_to_convert > 0:
                    base_vacation_candidates = employees_on_vacation[(employees_on_vacation['base_status'] == 'vacation') & (employees_on_vacation['final_status'] == 'vacation')]
                    if base_vacation_candidates.shape[0] >= remaining_to_convert:
                        indices_to_convert_base = np.random.choice(base_vacation_candidates.index, remaining_to_convert, replace=False)
                        employee_schedule_df.loc[indices_to_convert_base, 'final_status'] = 'working'
                    elif base_vacation_candidates.shape[0] > 0:
                         # Convert all remaining base vacation days if not enough to meet the requirement
                        indices_to_convert_base = base_vacation_candidates.index
                        employee_schedule_df.loc[indices_to_convert_base, 'final_status'] = 'working'


display(employee_schedule_df.head())
display(employee_schedule_df.tail())
```

```python
# Determine Movement Types

# Initialize cell_usage_df
cell_usage_df = pd.DataFrame(columns=['event_type_category'])

# Determine Movement Types

total_movements = 1000000 # Total number of movements, assuming 1 million as before
move_x_y_count = int(total_movements * 0.14) # 14% move between any two cells (not cell 1)
move_from_1_count = int(total_movements * 0.42) # 42% move from cell 1 to any cell with type "move"
return_from_1_count = int(total_movements * 0.01) # 1% move from cell 1 to any cell with type "return"
ship_to_1_count = total_movements - move_x_y_count - move_from_1_count - return_from_1_count # Remaining 43% from any cell to cell 1 with type "ship"

# Create a list of movement types based on the counts
movement_types = (
    ['move_x_y'] * move_x_y_count +
    ['move_from_1'] * move_from_1_count +
    ['return_from_1'] * return_from_1_count +
    ['ship_to_1'] * ship_to_1_count
)

# Shuffle the list to randomize the distribution of movement types
np.random.shuffle(movement_types)

# Add the movement types to the DataFrame
cell_usage_df['event_type_category'] = movement_types

display(cell_usage_df.head())
display(cell_usage_df['event_type_category'].value_counts())
```

```python
# Select Warehouses, Sections, Rows, and Cells

# Create a mapping from cell_id to warehouse_id for easy lookup
cell_to_warehouse_map = warehouse_df.set_index('cell_id')['warehouse_id'].to_dict()

# Get valid cell_ids (excluding cell 1)
valid_cell_ids = warehouse_df[warehouse_df['cell_id'] > 1]['cell_id']

# Initialize lists to store generated data
cell_from_list = []
cell_to_list = []
warehouse_list = []

# Use vectorized operations based on movement type
move_from_1_mask = cell_usage_df['event_type_category'] == 'move_from_1'
return_from_1_mask = cell_usage_df['event_type_category'] == 'return_from_1'
ship_to_1_mask = cell_usage_df['event_type_category'] == 'ship_to_1'
move_x_y_mask = cell_usage_df['event_type_category'] == 'move_x_y'

# Handle 'move_from_1' and 'return_from_1'
cell_usage_df.loc[move_from_1_mask | return_from_1_mask, 'cell_from'] = 1
cell_usage_df.loc[move_from_1_mask | return_from_1_mask, 'cell_to'] = np.random.choice(valid_cell_ids, size=(move_from_1_mask | return_from_1_mask).sum())

# Handle 'ship_to_1'
cell_usage_df.loc[ship_to_1_mask, 'cell_from'] = np.random.choice(valid_cell_ids, size=ship_to_1_mask.sum())
cell_usage_df.loc[ship_to_1_mask, 'cell_to'] = 1

# Handle 'move_x_y'
num_move_x_y = move_x_y_mask.sum()
random_warehouses = np.random.choice(warehouse_df['warehouse_id'].unique(), size=num_move_x_y)

cell_from_move_x_y = []
cell_to_move_x_y = []
warehouse_move_x_y = []

for warehouse_id in warehouse_df['warehouse_id'].unique():
    warehouse_mask = random_warehouses == warehouse_id
    num_in_warehouse = warehouse_mask.sum()

    if num_in_warehouse > 0:
        valid_cells_in_warehouse = warehouse_df[(warehouse_df['warehouse_id'] == warehouse_id) & (warehouse_df['cell_id'] > 1)]['cell_id']

        if len(valid_cells_in_warehouse) >= 2:
            # Select two distinct random cells within the warehouse
            selected_cells = np.random.choice(valid_cells_in_warehouse, size=num_in_warehouse * 2, replace=True).reshape(-1, 2) # Use replace=True for simplicity with large numbers
            cell_from_move_x_y.extend(selected_cells[:, 0])
            cell_to_move_x_y.extend(selected_cells[:, 1])
            warehouse_move_x_y.extend([warehouse_id] * num_in_warehouse)
        elif len(valid_cells_in_warehouse) == 1 and num_in_warehouse > 0:
             # If only one valid cell, assign that cell to both from and to
            cell_from_move_x_y.extend([valid_cells_in_warehouse.iloc[0]] * num_in_warehouse)
            cell_to_move_x_y.extend([valid_cells_in_warehouse.iloc[0]] * num_in_warehouse)
            warehouse_move_x_y.extend([warehouse_id] * num_in_warehouse)
        else:
            # Handle cases with no valid cells in a warehouse
            cell_from_move_x_y.extend([np.nan] * num_in_warehouse)
            cell_to_move_x_y.extend([np.nan] * num_in_warehouse)
            warehouse_move_x_y.extend([np.nan] * num_in_warehouse)


# Assign the generated cell_from, cell_to, and warehouse_id for 'move_x_y' movements
cell_usage_df.loc[move_x_y_mask, 'cell_from'] = cell_from_move_x_y
cell_usage_df.loc[move_x_y_mask, 'cell_to'] = cell_to_move_x_y
cell_usage_df.loc[move_x_y_mask, 'warehouse_id'] = warehouse_move_x_y

# Fill warehouse_id for 'move_from_1', 'return_from_1', and 'ship_to_1' using the map
cell_usage_df.loc[move_from_1_mask | return_from_1_mask, 'warehouse_id'] = cell_usage_df.loc[move_from_1_mask | return_from_1_mask, 'cell_to'].map(cell_to_warehouse_map)
cell_usage_df.loc[ship_to_1_mask, 'warehouse_id'] = cell_usage_df.loc[ship_to_1_mask, 'cell_from'].map(cell_to_warehouse_map)


display(cell_usage_df.head())
display(cell_usage_df.tail())
```

```python
# Load emart table
# batch load
def load_dataframe_to_bigquery(df: pd.DataFrame, table_name: str, project_id: str, chunksize=100000):
    total_rows = len(df)
    num_batches = (total_rows + chunksize - 1) // chunksize  # Ceiling division to get the total number of batches

    # Loop over batches and load to BigQuery
    for i in tqdm(range(num_batches), desc="Loading to BigQuery", unit="batch", leave=True):
        start_idx = i * chunksize
        end_idx = min((i + 1) * chunksize, total_rows)
        batch_df = df.iloc[start_idx:end_idx].copy() # Create a copy to avoid modifying the original DataFrame

        # Convert timestamp column to microsecond precision
        for col in batch_df.select_dtypes(include=['datetime64[ns]']).columns:
            batch_df[col] = batch_df[col].astype('datetime64[us]')


        # Load batch_df to BigQuery
        pandas_gbq.to_gbq(batch_df, table_name, project_id=project_id, if_exists='append')

load_dataframe_to_bigquery(
    df=cell_usage_df,
    table_name='ecom_data_mart.cell_usage',
    project_id='kaggle-bq-249420',
    )
```
